# Введение в Jupyter

Вы находитесь в Jupyter (а точнее, в его продвинутой версии JupyterLab) - интерактивном редакторе Python кода, основном компоненте ViXtract. В первом уроке вы познакомитесь с основными принципами его работы. Кликните мышкой на эту строку и нажимайте `Shift`+`Enter`, чтобы последовательно выполнять ячейки и продвигаться по уроку.

Перед вами тетрадка формата ipynb. В ближайшие несколько минут вы узнаете про основные возможности работы с подобными тетрадками.  

Первое, на что стоит обратить внимание - **ячейки**, они же главные элементы любой тетрадки. **Jupyter** поддерживает работу со многими языками (среди наиболее популярных - **Python** и **R**), но нас интересуют две основные возможности ячейки:  
1. **Исполнение кода** - основное действие, выполняемое в любой тетрадке, стандартный режим ячейки при её создании, вы наверняка будете создавать, менять и удалять их ещё не раз. Для перехода в режим исполнения кода на активной ячейке нужно нажать `Y`  
2. **Режим Markdown** - применяется для написания текстовых комментариев, выводов, дополнительных сведений о выполняемой работе, при выполнении ячейки применяет **Markdown**-форматирование. Для перехода в режим форматирования текста на активной ячейке нужно нажать `M`

Исполнение кода в любой ячейке можно выполнить следующим образом:  
1. Кнопкой `▶` на панели инструментов в верхней части страницы  
1. Сочетанием клавиш `Ctrl`+`Enter` (выполняется на месте)  
1. Сочетанием клавиш `Alt`+`Enter` (после выполненной ячейки появляется новая ячейка, которая становится активной)  
1. Сочетанием клавиш `Shift`+`Enter` (ячейка выполняется, активной становится следующая ячейка, если такая есть, создается новая, если следующей ячейки нет)

````python
# Выполните код в этой и следующей ячейках сочетанием клавиш Shift+Enter. 
# Результат выполнения команды появляется сразу под ячейкой.
print('Ячейка выполнила код')
````

````python
# Ошибки тоже отображаются сразу под ячейкой, это очень удобно для отладки. 
# Выполнение этой ячейки в первый раз приведёт к ошибке - 
# так и должно быть, потому что в конце не поставлена закрывающая скобка.
print('Ячейка выполнила код'
````

На этом примере вы увидели еще одно преимущество Jupyter - возможность выполнять ячейки с кодом в любом порядке, при этом состояние программы сохраняется. Рассмотрим также возможности JupyterLab. Слева мы видим навигационную панель. В первую очередь нас интересует вкладка браузера файлов.  

Браузер файлов и меню File позволяют открывать, создавать, удалять, переименовывать, скачивать, копировать и делиться файлами и директориями. Файлы открываются стандартно – двойными кликом в браузере файлов из боковой панели.

Здесь мы можем создать новую тетрадку или текстовый файл, для чего необходимо нажать на `+`. Новый файл создается с именем по умолчанию. Переименовывается файл из контекстного меню браузера файлов или вкладки, на которой открыт файл.

Кнопка со стрелкой вверх позволяет загрузить файлы в текущую активную папку. Файлы также могут быть загружены в текущую директорию переносом их в панель браузера файлов.
