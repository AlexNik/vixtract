2. **–†–µ–∂–∏–º Markdown** - –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –¥–ª—è –Ω–∞–ø–∏—Å–∞–Ω–∏—è —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤, –≤—ã–≤–æ–¥–æ–≤, –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–≤–µ–¥–µ–Ω–∏–π –æ –≤—ã–ø–æ–ª–Ω—è–µ–º–æ–π —Ä–∞–±–æ—Ç–µ, –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ —è—á–µ–π–∫–∏ –ø—Ä–∏–º–µ–Ω—è–µ—Ç **Markdown**-—Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ. –î–ª—è –ø–µ—Ä–µ—Ö–æ–¥–∞ –≤ —Ä–µ–∂–∏–º —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ –Ω–∞ –∞–∫—Ç–∏–≤–Ω–æ–π —è—á–µ–π–∫–µ –Ω—É–∂–Ω–æ –Ω–∞–∂–∞—Ç—å `M`

–ò—Å–ø–æ–ª–Ω–µ–Ω–∏–µ –∫–æ–¥–∞ –≤ –ª—é–±–æ–π —è—á–µ–π–∫–µ –º–æ–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:  
1. –ö–Ω–æ–ø–∫–æ–π `‚ñ∂` –Ω–∞ –ø–∞–Ω–µ–ª–∏ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–æ–≤ –≤ –≤–µ—Ä—Ö–Ω–µ–π —á–∞—Å—Ç–∏ —Å—Ç—Ä–∞–Ω–∏—Ü—ã  
1. –°–æ—á–µ—Ç–∞–Ω–∏–µ–º –∫–ª–∞–≤–∏—à `Ctrl`+`Enter` (–≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –Ω–∞ –º–µ—Å—Ç–µ)  
1. –°–æ—á–µ—Ç–∞–Ω–∏–µ–º –∫–ª–∞–≤–∏—à `Alt`+`Enter` (–ø–æ—Å–ª–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω–æ–π —è—á–µ–π–∫–∏ –ø–æ—è–≤–ª—è–µ—Ç—Å—è –Ω–æ–≤–∞—è —è—á–µ–π–∫–∞, –∫–æ—Ç–æ—Ä–∞—è —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –∞–∫—Ç–∏–≤–Ω–æ–π)  
1. –°–æ—á–µ—Ç–∞–Ω–∏–µ–º –∫–ª–∞–≤–∏—à `Shift`+`Enter` (—è—á–µ–π–∫–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –∞–∫—Ç–∏–≤–Ω–æ–π —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Å–ª–µ–¥—É—é—â–∞—è —è—á–µ–π–∫–∞, –µ—Å–ª–∏ —Ç–∞–∫–∞—è –µ—Å—Ç—å, —Å–æ–∑–¥–∞–µ—Ç—Å—è –Ω–æ–≤–∞—è, –µ—Å–ª–∏ —Å–ª–µ–¥—É—é—â–µ–π —è—á–µ–π–∫–∏ –Ω–µ—Ç)

````python
# –í—ã–ø–æ–ª–Ω–∏—Ç–µ –∫–æ–¥ –≤ —ç—Ç–æ–π –∏ —Å–ª–µ–¥—É—é—â–µ–π —è—á–µ–π–∫–∞—Ö —Å–æ—á–µ—Ç–∞–Ω–∏–µ–º –∫–ª–∞–≤–∏—à
print(''.join(reversed('–¥1–æ2–∫3 4–∞5–ª6–∏7–Ω8–ª9–æ0–ø1—ã2–≤3 4–∞5–∫6–π7–µ8—á9–Ø0'[::2])))
````

````python
# –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —ç—Ç–æ–π —è—á–µ–π–∫–∏ –≤ –ø–µ—Ä–≤—ã–π —Ä–∞–∑ –ø—Ä–∏–≤–µ–¥—ë—Ç –∫ –æ—à–∏–±–∫–µ - —Ç–∞–∫ –∏ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å
print(text)
````

````python
–í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —ç—Ç–æ–π —è—á–µ–π–∫–∏ –≤ —Ä–µ–∂–∏–º–µ —Ä–∞–±–æ—Ç—ã —Å –∫–æ–¥–æ–º –ø—Ä–∏–≤–µ–¥—ë—Ç –∫ **–æ—à–∏–±–∫–µ**. –ö —Ç–æ–º—É –∂–µ, –Ω–µ –±—É–¥—É—Ç –ø—Ä–∏–º–µ–Ω–µ–Ω–æ *—Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ*. –ù—É–∂–Ω–æ —Å–º–µ–Ω–∏—Ç—å —Ç–∏–ø —è—á–µ–π–∫–∏ –∫–ª–∞–≤–∏—à–µ–π `M`, –∏ –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Å–Ω–æ–≤–∞, —á—Ç–æ–±—ã —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–º–µ–Ω–∏–ª–æ—Å—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ.
````

----------

–í–æ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –º–æ–≥—É—Ç —Ç–∞–∫–∂–µ –ø—Ä–∏–≥–æ–¥–∏—Ç—å—Å—è —Å–ª–µ–¥—É—é—â–∏–µ –≥–æ—Ä—è—á–∏–µ –∫–ª–∞–≤–∏—à–∏:  
- `B` —Å–æ–∑–¥–∞—ë—Ç —è—á–µ–π–∫—É –Ω–∞–¥ –∞–∫—Ç–∏–≤–Ω–æ–π  
- `A` —Å–æ–∑–¥–∞—ë—Ç —è—á–µ–π–∫—É –ø–æ–¥ –∞–∫—Ç–∏–≤–Ω–æ–π  
- `D`, –Ω–∞–∂–∞—Ç–∞—è –¥–≤–∞–∂–¥—ã, —É–¥–∞–ª—è–µ—Ç –∞–∫—Ç–∏–≤–Ω—É—é —è—á–µ–π–∫—É  
- `Z` –æ—Ç–º–µ–Ω—è–µ—Ç —É–¥–∞–ª–µ–Ω–∏–µ —è—á–µ–π–∫–∏

–°–æ–∑–¥–∞–¥–∏–º —è—á–µ–π–∫—É *–Ω–∞–¥ —Ç–µ–∫—É—â–µ–π* –∏ –≤—ã–ø–æ–ª–Ω–∏–º –≤ –Ω–µ–π —Å–ª–µ–¥—É—é—â–∏–π –∫–æ–¥:

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ç–∞–∫–∂–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ JupyterLab. –°–ª–µ–≤–∞ –º—ã –≤–∏–¥–∏–º –Ω–∞–≤–∏–≥–∞—Ü–∏–æ–Ω–Ω—É—é –ø–∞–Ω–µ–ª—å. –í –ø–µ—Ä–≤—É—é –æ—á–µ—Ä–µ–¥—å –Ω–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç –≤–∫–ª–∞–¥–∫–∞ –±—Ä–∞—É–∑–µ—Ä–∞ —Ñ–∞–π–ª–æ–≤ - üñø.  

–ë—Ä–∞—É–∑–µ—Ä —Ñ–∞–π–ª–æ–≤ –∏ –º–µ–Ω—é File –ø–æ–∑–≤–æ–ª—è—é—Ç –æ—Ç–∫—Ä—ã–≤–∞—Ç—å, —Å–æ–∑–¥–∞–≤–∞—Ç—å, —É–¥–∞–ª—è—Ç—å, –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞—Ç—å, —Å–∫–∞—á–∏–≤–∞—Ç—å, –∫–æ–ø–∏—Ä–æ–≤–∞—Ç—å –∏ –¥–µ–ª–∏—Ç—å—Å—è —Ñ–∞–π–ª–∞–º–∏ –∏ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è–º–∏. –§–∞–π–ª—ã –æ—Ç–∫—Ä—ã–≤–∞—é—Ç—Å—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ ‚Äì –¥–≤–æ–π–Ω—ã–º–∏ –∫–ª–∏–∫–æ–º –≤ –±—Ä–∞—É–∑–µ—Ä–µ —Ñ–∞–π–ª–æ–≤ –∏–∑ –±–æ–∫–æ–≤–æ–π –ø–∞–Ω–µ–ª–∏.

–ó–¥–µ—Å—å –º—ã –º–æ–∂–µ–º —Å–æ–∑–¥–∞—Ç—å –Ω–æ–≤—É—é —Ç–µ—Ç—Ä–∞–¥–∫—É –∏–ª–∏ —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª, –¥–ª—è —á–µ–≥–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–∂–∞—Ç—å –Ω–∞ `+`. –ù–æ–≤—ã–π —Ñ–∞–π–ª —Å–æ–∑–¥–∞–µ—Ç—Å—è —Å –∏–º–µ–Ω–µ–º –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é. –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ—Ç—Å—è —Ñ–∞–π–ª –∏–∑ –∫–æ–Ω—Ç–µ–∫—Å—Ç–Ω–æ–≥–æ –º–µ–Ω—é –±—Ä–∞—É–∑–µ—Ä–∞ —Ñ–∞–π–ª–æ–≤ –∏–ª–∏ –≤–∫–ª–∞–¥–∫–∏, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–π –æ—Ç–∫—Ä—ã—Ç —Ñ–∞–π–ª.

–ö–Ω–æ–ø–∫–∞ —Å–æ —Å—Ç—Ä–µ–ª–∫–æ–π –≤–≤–µ—Ä—Ö –ø–æ–∑–≤–æ–ª—è–µ—Ç –∑–∞–≥—Ä—É–∑–∏—Ç—å —Ñ–∞–π–ª—ã –≤ —Ç–µ–∫—É—â—É—é –∞–∫—Ç–∏–≤–Ω—É—é –ø–∞–ø–∫—É. –§–∞–π–ª—ã —Ç–∞–∫–∂–µ –º–æ–≥—É—Ç –±—ã—Ç—å –∑–∞–≥—Ä—É–∂–µ–Ω—ã –≤ —Ç–µ–∫—É—â—É—é –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é –ø–µ—Ä–µ–Ω–æ—Å–æ–º –∏—Ö –≤ –ø–∞–Ω–µ–ª—å –±—Ä–∞—É–∑–µ—Ä–∞ —Ñ–∞–π–ª–æ–≤.

-----------

Jupyter - –Ω–µ –ø—Ä–æ—Å—Ç–æ —Ç–µ—Ç—Ä–∞–¥–∫–∞ —Å –∫–æ–¥–æ–º –∏ —Ç–µ–∫—Å—Ç–æ–º. –ü–æ–∫–∞ –≤—ã –Ω–µ –ø–µ—Ä–µ–∑–∞–ø—É—Å—Ç–∏—Ç–µ —è–¥—Ä–æ, –≤—Å–µ –∫–æ–≥–¥–∞-–ª–∏–±–æ –∑–∞–¥–µ–π—Å—Ç–≤–æ–≤–∞–Ω–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –±—É–¥—É—Ç –¥–æ—Å—Ç—É–ø–Ω—ã –≤ –ª—é–±–æ–º –º–µ—Å—Ç–µ —Ç–µ—Ç—Ä–∞–¥–∫–∏. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø—Ä–∏–º–µ—Ä.  

–°–æ–∑–¥–∞–π—Ç–µ —è—á–µ–π–∫—É –Ω–∏–∂–µ, –≤–≤–µ–¥–∏—Ç–µ `text = 'example'`, –∏—Å–ø–æ–ª–Ω–∏—Ç–µ, –∞ –∑–∞—Ç–µ–º —É–¥–∞–ª–∏—Ç–µ —è—á–µ–π–∫—É. –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –≤–µ—Ä–Ω–∏—Ç–µ—Å—å –∫ –Ω–∞—á–∞–ª—É —Ç–µ—Ç—Ä–∞–¥–∫–∏ –∏ –≤—ã–ø–æ–ª–Ω–∏—Ç–µ —è—á–µ–π–∫—É —Å –æ—à–∏–±–∫–æ–π –µ—â—ë —Ä–∞–∑.

----------

–ü–æ—Å–ª–µ–¥–Ω–µ–µ, –Ω–æ —Ç–æ–∂–µ –≤–∞–∂–Ω–æ–µ - –ø–æ–¥—Å–∫–∞–∑–∫–∏ –ø—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å –∫–æ–¥–æ–º. –ò—Å–ø–æ–ª—å–∑—É—è —Ä–∞–∑–ª–∏—á–Ω—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∏ –∏—Ö —Ñ—É–Ω–∫—Ü–∏–∏ —Å–ª–æ–∂–Ω–æ –∑–∞–ø–æ–º–Ω–∏—Ç—å –≤—Å–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç—å. –î–ª—è —ç—Ç–æ–≥–æ —É–¥–æ–±–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–æ—á–µ—Ç–∞–Ω–∏–µ `Shift`+`Tab`, –∫–æ–≥–¥–∞ –∫—É—Ä—Å–æ—Ä –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ —Å–∫–æ–±–∫–∞—Ö –≤—ã–ø–æ–ª–Ω—è–µ–º–æ–π —Ñ—É–Ω–∫—Ü–∏–∏.

````python
import numpy as np
````

````python
# –ü–µ—Ä–µ–º–µ—â–∞–µ–º –∫—É—Ä—Å–æ—Ä –≤ —Å–∫–æ–±–∫–∏, –Ω–∞–∂–∏–º–∞–µ–º Shift+Tab
np.matmul()
````

````python
# –ù–∞–∂–∏–º–∞–µ–º Tab, –Ω–µ –¥–æ–ø–∏—Å—ã–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—é, –∏ –≤–∏–¥–∏–º –¥–æ—Å—Ç—É–ø–Ω—ã–µ –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –¥–∞—Ç–∞–º–∏-–≤—Ä–µ–º–µ–Ω–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –º–æ–¥—É–ª–∏
np.datetime_
````
